```python
# IMPORT MODULES
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# IMPORT DATA AND TRANSFORM TO DATAFRAME
pwd = os.getcwd()
playstore = pd.read_csv(pwd + '/datasets/googleplaystore.csv')
playstore.head()
```

```python
playstore.info()
```

```python
# CLEANING THE DATA
# remove duplicate apps
apps = playstore.drop_duplicates()
apps.info()
```

```python
# removing unwanted characters from installs and price
# 1. create list of unwanted characters
unwanted = ['+',',','$']
# list of columns to clean
cols2clean = ['Installs', 'Price']
# loop to clean
for c in cols2clean:
    for u in unwanted:
        apps[c] = apps[c].apply(lambda x: x.replace(u, ''))
apps.head()
```

```python
# EXPLORE THE DATA
# count categories
categories = len(apps['Category'].unique())
print(categories, 'Total categories')
```

```python
# count total number of apps in each category
appsincat = apps['Category'].value_counts()
print(appsincat)
```

```python
#  Distribution of app ratings
rating_mean = apps['Rating'].mean()
print(round(rating_mean,1), 'is the average app rating')
```
